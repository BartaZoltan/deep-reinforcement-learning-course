# Session 3 Asset Builder

This notebook generates simple FrozenLake trajectory assets from manually specified action sequences.

Action encoding:
- `0`: LEFT
- `1`: DOWN
- `2`: RIGHT
- `3`: UP


```python
from pathlib import Path
import importlib.util

candidate_paths = [
    Path("utils.py"),
    Path("notebooks/sessions/3_monte_carlo_methods/utils.py"),
]

utils_path = next((p.resolve() for p in candidate_paths if p.exists()), None)
if utils_path is None:
    raise FileNotFoundError("Could not locate Session 3 utils.py")

spec = importlib.util.spec_from_file_location("mc_utils", utils_path)
mc_utils = importlib.util.module_from_spec(spec)
assert spec.loader is not None
spec.loader.exec_module(mc_utils)

export_frozenlake_manual_trajectory = mc_utils.export_frozenlake_manual_trajectory
```

```python
hole_run = export_frozenlake_manual_trajectory(
    actions=[1, 1, 2, 2, 2],
    output_dir=Path("assets/manual_trajectories/hole_run"),
    is_slippery=False,
    reset_seed=0,
    gif_name="frozenlake_hole_run.gif",
    frame_prefix="hole",
    fps=1.0,
)
hole_run
```

```python
goal_run = export_frozenlake_manual_trajectory(
    actions=[1, 1, 2, 2, 1, 2],  # DOWN, DOWN, RIGHT, DOWN, RIGHT, RIGHT -> reaches goal
    output_dir=Path("assets/manual_trajectories/goal_run"),
    is_slippery=False,
    reset_seed=0,
    gif_name="frozenlake_goal_run.gif",
    frame_prefix="goal",
    fps=1.0,
)
goal_run
```
