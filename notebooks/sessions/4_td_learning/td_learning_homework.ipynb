### Environment summary

In `Taxi-v3`, the agent must:
- navigate on a grid,
- pick up a passenger at the correct location,
- and drop the passenger off at the correct destination.

The action space is:
- `0`: south,
- `1`: north,
- `2`: east,
- `3`: west,
- `4`: pickup,
- `5`: dropoff.

The reward structure encourages efficient completion:
- small negative reward per step,
- larger penalties for illegal pickup/dropoff,
- positive reward for successful delivery.

## Minimal helpers

The helper functions below are provided so you can focus on the Q-learning logic itself.


```python
def running_mean(values, window: int = 50):
    values = np.asarray(values, dtype=float)
    if len(values) == 0:
        return values
    out = np.zeros_like(values)
    for i in range(len(values)):
        start = max(0, i - window + 1)
        out[i] = np.mean(values[start:i + 1])
    return out


def plot_return_curve(values, title: str, ylabel: str = 'Return'):
    x = np.arange(1, len(values) + 1)
    plt.figure(figsize=(7, 4))
    plt.plot(x, values, linewidth=2)
    plt.title(title)
    plt.xlabel('Episode index')
    plt.ylabel(ylabel)
    plt.grid(True, alpha=0.3)
    plt.tight_layout()
    plt.show()


def plot_curve_collection(x, curves, labels, title: str, xlabel: str, ylabel: str):
    plt.figure(figsize=(7.5, 4.2))
    for y, label in zip(curves, labels):
        plt.plot(x, y, linewidth=2, label=label)
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.grid(True, alpha=0.3)
    plt.legend()
    plt.tight_layout()
    plt.show()
```
